```python
#library needed
import sympy as sp
from sympy.logic.boolalg import truth_table
from itertools import product
```

```python
#Exercice 1
def implic(a,b):
    if a and not b :
        return False
    else :
        return True
def equiv(a,b):
    return a==b
def neg(x):
    return not x
def conj(a,b):
    return a and b
def disj(a,b):
    return a or b
a,b=True,True
print("a=true b=true : a->b",implic(a,b),"a<->b",equiv(a,b),"a^b",conj(a,b),"a v b ",disj(a,b))
a,b =True , False
print("a=true b=false : a->b",implic(a,b),"a<->b",equiv(a,b),"a^b",conj(a,b),"a v b ",disj(a,b))
a,b=False,True
print("a=false b=true : a->b",implic(a,b),"a<->b",equiv(a,b),"a^b",conj(a,b),"a v b ",disj(a,b))
a,b=False,False
print("a=false b=false : a->b",implic(a,b),"a<->b",equiv(a,b),"a^b",conj(a,b),"a v b ",disj(a,b))
#Exercice 2
```

```python
#Exercice 2
P=sp.symbols('P')
Q=sp.symbols('Q')
R=sp.symbols('R')
E=sp.And(sp.Implies(P,Q),sp.Implies(Q,R))



#tableau de verité de l'expression E
print("le table de verité est : ")
print(" P   Q   R |(P=>Q)^(Q=>R)")
t=truth_table(E,[P,Q,R])
for row in t :
    print(row)



#verifier la nature du cette expression
E_simplified=sp.simplify(E)

print("l'expression simplifiée est : ",E_simplified)

if E_simplified == True :
    print("l'expression est tautologie")

elif E_simplified == False :
    print("l'expression est contradiction")

else :
    print("l'expression n'est ni tautologie ni contradiction(contingence)")
```

```python
#Exercice 3
A=sp.symbols('A')
B=sp.symbols('B')
C=sp.symbols('C')
v_poss=[True,False]
comb=list(product(v_poss,repeat=2))
print("les combinaisons possibles sont : ")
print("  A  |  B  | not(A)  | A^B   | AvB  |  A=>B  | A<=>B")
print("-"*50)
for v in comb :
  print(v , "|",neg(v[0]), "|  ",conj(v[0],v[1]),"|  ",disj(v[0],v[1]) , "|  ",implic(v[0],v[1]),"|  ",equiv(v[0],v[1]))
```

```python
#Exercice 4
def is_taut(E):
  E_simplified=sp.simplify(E)
  return E_simplified == True
A=sp.symbols('A')
B=sp.symbols('B')
E1=sp.Equivalent(sp.Implies(A,B),sp.Or(neg(A),B))
E2=sp.Implies(sp.And(A,B),A)
E3=sp.Implies(A,sp.Or(A,B))
l=[E1,E2,E3]
for e in l :
  if is_taut(e) :
    print("l'expression ",e,"est tautologie")
  else :
    print("l'expression ",e,"n'est pas tautologie")

```
