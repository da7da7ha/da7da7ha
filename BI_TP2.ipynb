```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path='/content/drive/MyDrive/SuperMarketAnalysis.csv'
df=pd.read_csv(path)
```

```python
#KPIs
chiffre_affaire_total = df['Sales'].sum()
profit_total= df['gross income'].sum()
nombre_vente =df.shape[0]
vente_categorie=df.groupby('Product line')['Sales'].sum()
vente_city=df.groupby('City')['Sales'].sum()
print("Chiffre d'affaire total:",chiffre_affaire_total)
print()
print("Profit total:",profit_total)
print()
print("Nombre de vente:",nombre_vente)
print()
print("Vente par categorie:",vente_categorie)
print()
print("Vente par ville:",vente_city)
```

```python
#kpi:ventes par gategorie
df.groupby('Product line')['Sales'].sum().plot(kind='bar',title='Vente par categorie')
plt.show()
```

```python
#kpi : ventes par ville
df.groupby('City')['Sales'].sum().plot(kind='bar',title='Vente par ville')
plt.show()
```
